```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# loading the diabetics dataset to a pandas dataframe
diabetics_dataset = pd.read_csv('/content/drive/MyDrive/ML Project /diabetes2.csv')
diabetics_dataset
```

```python
diabetics_dataset['Outcome'].value_counts()
```

```python
# Separating the Data and Labels
x = diabetics_dataset.drop(columns='Outcome', axis=1)
y = diabetics_dataset['Outcome']
print(x)
print("-"*100)
print("Outcome Column")
print(y)
```

# **Data Standarization**

```python
scaler = StandardScaler()
```

```python
standarize_data = scaler.fit_transform(x)
print(standarize_data)
```

```python
x = standarize_data
y = diabetics_dataset['Outcome']
```

```python
print(x)
print(y)
```

# **Train_Test_Split**

```python
# Fixed the comma between X_train, X_test and Y_train, Y_test
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

# **Training the Model**

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the Support Vector Machine Classifier
classifier.fit(X_train, Y_train)
```

# **Model Evaluotion**

# **Accuracy Score**

```python
predictions = classifier.predict(X_test)
accuracy = accuracy_score(predictions, Y_test)
print(accuracy)
```
